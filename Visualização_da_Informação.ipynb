```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np


labels = ['Furtos', 'Roubo', 'Homicidio', 'Lesão Corporal']
primeiro_means = [226556, 16427, 448, 45531]
segundo_means = [164444, 141530, 422, 36769]

x = np.arange(len(labels))  
width = 0.35  

fig, ax = plt.subplots()
rects1 = ax.bar(x - width/2, primeiro_means, width, label='2020')
rects2 = ax.bar(x + width/2, segundo_means, width, label='2019')

ax.set_ylabel('quantidade')
ax.set_title('Crimes São Paulo Capital')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()


def autolabel(rects):
  
    for rect in rects:
        height = rect.get_height()
        ax.annotate('{}'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 1),  
                    textcoords="offset points",
                    ha='center', va='bottom')


autolabel(rects1)
autolabel(rects2)

fig.tight_layout()

plt.show()
```
